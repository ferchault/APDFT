```python
# plot coefficient of contracted 1s
coeff_1s_1 = []
for i in [b_sto, c_sto, n_sto, o_sto, f_sto]:
    coeff_1s_1.append(i[0][1][1])
```

```python
plt.plot([5,6,7,8,9],np.abs(np.array(coeff_1s_1)))
```

```python
from scipy.interpolate import CubicSpline
```

```python
ip = CubicSpline(np.array([5,6,7,8,9]),np.abs(np.array(coeff_1s_1)), bc_type='clamped')
```

```python
lsx = np.linspace(5, 9, 100)
plt.plot(lsx,ip.__call__(lsx))
```

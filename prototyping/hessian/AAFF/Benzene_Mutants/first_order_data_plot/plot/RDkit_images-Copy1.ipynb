```python
def drawMyMol(fname, myFontSize):
    d = rdMolDraw2D.MolDraw2DCairo(350, 300)
    d.SetFontSize(myFontSize)
    d.DrawMolecule(mol)
    d.FinishDrawing()
    d.WriteDrawingText(fname)
```

```python
for i in range(1,18):
    d = rdMolDraw2D.MolDraw2DCairo(400, 400 ,350,350 )
    d.SetFontSize(1.7)
    d.DrawMolecule(mols[i-1])
    d.FinishDrawing()
    d.WriteDrawingText("Mutants/{}.png".format(i))
```

```python
mg=Draw.MolsToGridImage(mols,molsPerRow=6,subImgSize=(200,300),legends=[str(x+1) for x in range(17)]) 
msvg=Draw.MolsToGridImage(mols,molsPerRow=6,subImgSize=(70,120),legends=[str(x+1) for x in range(17)],useSVG=True)
mg
```

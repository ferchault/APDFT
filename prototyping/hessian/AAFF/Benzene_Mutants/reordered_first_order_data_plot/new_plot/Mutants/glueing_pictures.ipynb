```python
import matplotlib.pyplot as plt
import matplotlib
from PIL import Image
```

```python
imgs=[]
for i in range(1,18):
    img = Image.open('./{}.png'.format(i))
    imgs.append(img)
```

```python
fig, axs = plt.subplots(6, 3, figsize=(24,48),gridspec_kw={"wspace":0})
axs = axs.flatten()
axs[1].set_title("Benzene B-N mutants",fontsize=80,pad=100)
i=1
for img, ax in zip(imgs, axs):
    ax.set_axis_off()
    ax.imshow(img)
    ax.text(140,420,str(i),size=50,\
            bbox=dict(facecolor='None', edgecolor='black', boxstyle='round',pad=.2))#,color="green")
    i+=1
axs[-1].set_axis_off()
plt.savefig("prova2.png",bbox_inc="tight")
```

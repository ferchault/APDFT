```python
fig,axs=plt.subplots(3,1,figsize=(7,8.3),gridspec_kw={"hspace":0.2})
axs[0].set_title("Energy predictions of \n benzene B-N mutants",fontsize=25)
axs[0].set_xlim(.95,3.05)
axs[0].plot(np.arange(1,4),dfs.opt[:3],lw=2.5,color="black")
axs[0].plot(np.arange(1,4),dfs.apdft3_vertical[:3],ls="--")
axs[0].scatter(np.arange(1,4),dfs.apdft3_vertical[:3])
axs[0].plot(np.arange(1,4),dfs.apdft3_relax[:3],lw=2,ls="--")
axs[0].scatter(np.arange(1,4),dfs.apdft3_relax[:3])
#two BN substitution
axs[1].set_xlim(3.8,14.2)
axs[1].plot(np.arange(4,15),dfs.opt[3:-3],lw=2.5,color="black")
axs[1].plot(np.arange(4,15),dfs.apdft3_vertical[3:-3],ls="--")
axs[1].scatter(np.arange(4,15),dfs.apdft3_vertical[3:-3])
axs[1].plot(np.arange(4,15),dfs.apdft3_relax[3:-3],lw=2,ls="--")
axs[1].scatter(np.arange(4,15),dfs.apdft3_relax[3:-3])
#three BN subs
axs[2].set_xlim(14.95,17.05)
axs[2].plot(np.arange(15,18),dfs.apdft3_vertical[14:17],ls="--",label="Vertical APDFT3")
axs[2].scatter(np.arange(15,18),dfs.apdft3_vertical[14:17],label=None)
axs[2].plot(np.arange(15,18),dfs.apdft3_relax[14:17],lw=2,ls="--",label="Relaxed APDFT3")
axs[2].scatter(np.arange(15,18),dfs.apdft3_relax[14:17],label=None)
axs[2].plot(np.arange(15,18),dfs.opt[14:17],lw=2.5,color="black",label="Target Energy")

#xticks
axs[0].set_xticks(np.arange(1,4))
axs[0].set_xticklabels(np.arange(1,4),fontsize=17)
axs[1].set_xticks(np.arange(4,15))
axs[1].set_xticklabels(np.arange(4,15),fontsize=17)
axs[2].set_xticks(np.arange(15,18))
axs[2].set_xticklabels(np.arange(15,18),fontsize=17)

#yticks 
axs[0].set_yticks([-234.225,-234.200,-234.175,-234.150,-234.125])
axs[0].set_yticklabels(["-234.225","-234.200","-234.175","-234.150","-234.125"],fontsize=16)
axs[1].set_yticklabels([np.around(x,decimals=2) for x in axs[1].get_yticks()],fontsize=16)
axs[2].set_yticklabels([np.around(x,decimals=2) for x in axs[2].get_yticks()],fontsize=16)

#axs[2].legend(loc=(-.25,3.5),fontsize=15,ncol=3)
axs[2].legend(loc=(-.05,-.7),fontsize=18,ncol=2,labelspacing=.1,columnspacing=1)

axs[1].set_ylabel("$E$ [Ha]",fontsize=30,labelpad=15)
plt.savefig("energy_plot.png",bbox_inches="tight")
```

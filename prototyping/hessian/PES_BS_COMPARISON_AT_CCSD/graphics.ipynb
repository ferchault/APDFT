```python
df3=pd.merge(dfprova,dfR_sto3G,on=['distance','ref','target'])
df3[(df3.target=='NN')&(df3.ref=='CO')]
```

```python
from bokeh.io import output_notebook, show 
from bokeh.plotting import figure
import ipyvolume.bokeh
from bokeh.models import HoverTool,ColumnDataSource
```

```python
p = figure(plot_width=1000, plot_height=700)

alcolor={0:'purple',1:'blue',2:'green',3:'yellow',4:'red'}
def alcol(ao):
    return alcolor[ao]
Alcol=[alcol(x) for x in list(df3.alchemy)]
df3['alcol']=Alcol
source = ColumnDataSource(df3)
p.scatter('delta_rho', 'abs_err', source=source,color='alcol')
hover = HoverTool()
hover.tooltips = [("Reference","@ref"), ("Target", "@target"),("Distance",'@distance'),('Alchemy','@alchemy')]
hover.mode = 'mouse'
p.add_tools(hover)
p.xaxis.axis_label = 'Time'
p.yaxis.axis_label = 'Stock Price'
show(p)
```

How much 2nd derivative finite difference changes with respect to the place taken?

```python
pes=[]
for R in np.linspace(1.97,2.06,10):
    pes.append([R,uhf(R,[7.,7.])])
```

```python
FDderiv=[]
i=1
while i<len(pes)-1:
    FDderiv.append([pes[i][0],to_cm((pes[i-1][1]-2*pes[i][1]+pes[i+1][1])*10000,Mu_NN)])
    i+=1
```

```python
plt.subplot(2,1,1)
plt.plot([x[0] for x in pes[1:-1]],[x[1] for x in pes[1:-1]])
plt.subplot(2,1,2)
plt.plot([x[0] for x in FDderiv],[x[1] for x in FDderiv])
```
